```python
#Cell 1
import pandas as pd

# students and test results
students = {
    'hours': [29, 9, 10, 38, 16, 26, 50, 10, 30, 33, 43, 2, 39, 15, 44, 29, 41, 15, 24, 50],
    'test_results': [65, 7, 8, 76, 23, 56, 100, 3, 74, 48, 73, 0, 62, 37, 74, 40, 90, 42, 58, 100]
}

student_data = pd.DataFrame(data=students)
```

```python
import numpy as np
import matplotlib.pyplot as plt

hours = student_data['hours']
test_results = student_data['test_results']

# Lin reg using polyfit fn with hours as x and test_results as y
coefficients = np.polyfit(hours, test_results, 1)

# get slope and intercept
slope, intercept = coefficients

# get equation
print(f"test_results = {slope:.2f} * hours + {intercept:.2f}")

# predict test results line and graph
predicted_results = slope * hours + intercept

# plot original data and the regression line
plt.scatter(hours, test_results, label="Original data", color="blue")
plt.plot(hours, predicted_results, label="Regression line", color="red")
plt.xlabel("hours studied")
plt.ylabel("test results")
plt.legend()
plt.title("hours studied to test results")
plt.show()
```

```python
# def for generating reg model

import csv
# Predicted House Price = 148.73 * Area - 18.85 * Bedrooms + 54.40

# data 
areas = []
num_beds = []
house_prices = []

# read csv
with open('houses.csv', mode='r') as file:
    csv_reader = csv.reader(file)
    next(csv_reader) 

    for row in csv_reader:
        areas.append(float(row[5]))  
        num_beds.append(float(row[4]))  
        house_prices.append(float(row[3]))  

areas = np.array(areas)
num_beds = np.array(num_beds)
house_prices = np.array(house_prices)
```

```python
# array for beds and area
X = np.column_stack((areas, num_beds, np.ones(len(areas))))

# house prices
y = house_prices

# linear reg using beds+area and prices
coefficients, residuals, rank, s = np.linalg.lstsq(X, y, rcond=None)

# find coefficients
w_area, w_beds, b = coefficients

print(f"price = {w_area:.2f} * area + {w_beds:.2f} * beds + {b:.2f}")

# solving the problem
predicted_price = w_area * 2000 + w_beds * 3 + b
print(f"{predicted_price:.2f}")



```
